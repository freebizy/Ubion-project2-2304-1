```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
a = pd.read_csv('./진짜코스닥내수수출.csv', encoding='cp949')
```

```python
a
```

```python
# 연결에 개별 채우기
connected_cols = [col for col in a.columns if "연결" in col]

for connected_col in connected_cols:
    original_col = connected_col.replace("연결", "별도")
    if original_col in a.columns:
        a.loc[a[connected_col].isnull(), connected_col] = a[original_col]
```

```python
내수수출= a[['회사명', '거래소코드', '회계년도', '매출유형구분(연결)', 'T_1기 수출금액(연결)',
       'T_1기 내수금액(연결)', 'T기 수출금액(연결)', 'T기 내수금액(연결)']]
```

```python
내수수출=내수수출[내수수출['매출유형구분(연결)']==90]
내수수출.head(10)
```

```python
내수수출=내수수출.drop_duplicates(keep='last')
```

```python
내수수출.shape
```

```python
내수수출.isna().sum()
```

```python
내수수출['내수변동'] = 내수수출.groupby(['회사명','거래소코드'])['T기 내수금액(연결)'].pct_change()*100
내수수출['수출변동'] = 내수수출.groupby(['회사명','거래소코드'])['T기 수출금액(연결)'].pct_change()*100
```

```python
내수수출['내수변동']=내수수출['내수변동'].fillna(0.0)
내수수출['내수변동'][0] = (15417000.0-11119000.0)/11119000.0 *100
내수수출['수출변동']=내수수출['수출변동'].fillna(0.0)
내수수출['수출변동'][0] = (371000.0-231000.0)/231000.0 *100
```

```python
내수수출.columns
```

```python
내수변화=[]
list = 내수수출['내수변동']

for i in list:
    if i > 0:
         내수변화.append('+')
    elif i < 0:
         내수변화.append('-')
    else:
         내수변화.append('0')

수출변화=[]
list = 내수수출['수출변동']

for i in list:
    if i > 0:
        수출변화.append('+')
    elif i < 0:
         수출변화.append('-')
    else:
         수출변화.append('0')
```

```python
내수수출['내수변화']=내수변화
내수수출['수출변화']=수출변화
```

```python
내수수출['T-1기 생산금액(연결)'] = 내수수출['T_1기 수출금액(연결)'] + 내수수출['T_1기 내수금액(연결)']
내수수출['T기 생산금액(연결)'] = 내수수출['T기 수출금액(연결)'] + 내수수출['T기 내수금액(연결)']
```

```python
내수수출['생산변동'] = 내수수출.groupby(['회사명','거래소코드'])['T기 생산금액(연결)'].pct_change()*100
```

```python
내수수출
```

```python
내수수출['생산변동'][0] = (15788000-11350000)/11350000 *100
내수수출['생산변동']=내수수출['생산변동'].fillna(0.0)
```

```python
생산변화=[]
list = 내수수출['생산변동']

for i in list:
    if i > 0:
        생산변화.append('+')
    elif i < 0:
         생산변화.append('-')
    else:
         생산변화.append('0')
```

```python
내수수출['생산변화']= 생산변화
내수수출['내수변화'].value_counts()
```

```python
내수수출['수출변화'].value_counts()
```

```python
내수수출['생산변화'].value_counts()
```

```python
내수수출.columns
```

```python
내수수출
```

```python
내수수출.drop(columns=['매출유형구분(연결)','T_1기 수출금액(연결)', 'T_1기 내수금액(연결)', 'T기 수출금액(연결)', 'T기 내수금액(연결)', '내수변동', '수출변동','T-1기 생산금액(연결)', 'T기 생산금액(연결)', '생산변동'], axis=1, inplace=True)
```
